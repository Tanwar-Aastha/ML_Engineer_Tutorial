## **FastAPI**

### What is API?
API -> `Application Programming Interface`<br>
2 or more software components communicate with each other with a set of rules, protocols etc.


**FastAPI**
_________
- High performance web framework for building APIs with python.
- It is build up on 2 Libraries:
    - `Starlette`: manage how the API receives the requests and sends back responses.
    - `Pydantic`: A **data validation library**. Checks if the data coming into the API is correct & in the right format.
- It is fast to code -> automatic input validation, auto-generated interactive documentation & seamless integration with modern ecosystem (ML/DL, Docker, SQL Alchemy etc.)

#### NOTE
To run the fastapi code:
```
uvicorn <file_name>:<FastAPIobject_name> --reload
```

`--reload` will help in auto-updating the changes while you are working on the code and can view the change at the same time.

## HTTP Methods

Based on user interaction we divide software into 2 types:
1. **STATIC -** user doesn't interact much.(are mostly used to receive the info)<br>
        for example - Calendar, Clock
2. **DYNAMIC -** user interact much (2 way communication b/w user and software)<br>
        for example - Microsoft Excel
___
- Client - software that requests information
- Server -  a computer or software that provides those services
- HTTP - protocol that helps in establishing the communication b/w client and server

___
___
***HTTP METHODS :***<br>

**GET -**  to retrieve and requesting something

**POST -** to send some info to website/server

**PUT -** to update some info

**DELETE -** to delete some info

___
**HTTP Codes:**
- `2xx`: Success
- `3xx`: Redirection (Needs more action)
- `4xx`: Client Error (Something wrong with the request from  the client)
- `5xx`: Server Error (Something wrong on the server side)

___
___
***HTTP Exception :***
It returns customer HTTP error response when something is wrong with the API.

### Path Parameters
- Dynamic Segments of a URL used to identify a specific resource.


The `Path()` function in FastAPI -

It provides
   - metadata,
   - validation rules and
   - documentation hints

for path parameters in the API end-points.

## Query Parameters

They are the optional key-value pairs appended to the end of a URL (pass additional data)
- It starts with `?`
- Multiple parameters are separated by `&`

Example - `/patients?city=Delhi&sort_by=age`

Here, `Query()` function helps in declaring, validating & documenting the API end-points.
