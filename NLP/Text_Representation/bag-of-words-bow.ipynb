```python
# importing libraries
import nltk
import re
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

ps = PorterStemmer()
# nltk.download("punkt")
# nltk.download("stopwords")
```

```python
df = pd.read_csv("/kaggle/input/datasets/hamnawaseem112222222/sms-spam-collection-5572-labeled-sms-messages/sms_spam_collection (1).csv")
df.head()
```

### Preprocessing the text

```python
corpus = []

for i in range(0, len(df)):
    # converting the text into lower case
    text = re.sub("[^a-zA-Z]", " ", df["message"][i])  # removing punctuations
    text = text.lower()
    # converting sentence into words
    text = text.split()
    text = [ps.stem(i) for i in text if i not in set(stopwords.words("english"))]  # removing stopwords and performing stemming
    message = " ".join(text)

    corpus.append(message)
    
    
    
```

### Performing BOW

```python
# applying bag of words

# creatign a vector object
vector = CountVectorizer(max_features=100, binary=True)

# fiting & transforming messages
X = vector.fit_transform(corpus).toarray()
```

### TF-IDF

```python
tfidf = TfidfVectorizer(max_features=100, ngram_range=(2,2))

X = tfidf.fit_transform(corpus).toarray()
```
