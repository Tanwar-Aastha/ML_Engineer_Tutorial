```python
import pandas as pd

data = {
    "text": ["The food is good", "The food is bad", "Pizza is Amazing"],
    "output": [1, 0, 1]
}

df = pd.DataFrame(data)

# Step 1: Create vocabulary
vocab = set()
for sentence in df["text"]:
    for word in sentence.split():
        vocab.add(word)

vocab = sorted(vocab)
print("Vocabulary:", vocab)

# Step 2: Create word index mapping
word_index = {word: i for i, word in enumerate(vocab)}

# Step 3: Encode sentences
def encode(sentence):
    vector = [0] * len(vocab)
    for word in sentence.split():
        if word in word_index:
            vector[word_index[word]] = 1
    return vector

df["encoded"] = df["text"].apply(encode)

print(df)
```
